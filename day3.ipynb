```python
result = 0
import numpy as np
with open('input3.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        battery_bank = [int(d) for d in str(int(line))]
        sorted_battery = sorted(battery_bank[:-1], reverse=True)
        highest_number = max(sorted_battery)
        highest_index = battery_bank.index(highest_number)
        second_highest_number = max(battery_bank[(highest_index + 1):])
        highest_joltage = int(str(highest_number) + str(second_highest_number))
        print(highest_joltage)
        result = result + highest_joltage

print(result)
```
