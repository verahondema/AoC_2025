```python
counter = 0
import re
with open('input1.txt', 'r') as file:
    instructions = file.readlines()
    #retrieve all instructions in separate lines
    dial = 50
    #the dial starts at 50
    for instruction in instructions:
        #read every instruction apart
        direction = instruction[0]
        #retrieve the direction from the instruction
        totalPoints = int(instruction[1:])
        #retrieve the amount of points one should turn the dial
        points = totalPoints - (int(totalPoints/100))*100
        #remove all complete rounds the dial turns
        if direction == 'L':
            #when the dial has to turn to the left
            dial = dial - points
            if dial == 0:
                counter = counter + 1
            elif dial < 0:
                dial = 100 - abs(dial)
        elif direction == 'R':
            dial = dial + points
            if dial == 100:
                counter = counter + 1
                dial = 0
            elif dial > 100:
                dial = dial - 100

print(f"Counter: {counter}")
```
