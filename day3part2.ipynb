```python
result = 0
import numpy as np
with open('input3.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        battery_bank = [int(d) for d in str(int(line))]
        sorted_battery = sorted(battery_bank[:-11], reverse=True)
        highest_number = max(sorted_battery)
        highest_index = battery_bank.index(highest_number)
        highest_joltage = highest_number
        for i in reversed(range(0, 11)):
            if i == 0:
                sorted_battery = sorted(battery_bank[(highest_index + 1):], reverse=True)
                highest_number = max(sorted_battery)
            else:
                sorted_battery = sorted(battery_bank[(highest_index + 1):-i], reverse=True)
                highest_number = max(sorted_battery)
                highest_index = battery_bank.index(highest_number, (highest_index + 1), -i)
            highest_joltage = int(str(highest_joltage) + str(highest_number))

        result = result + highest_joltage

print(result)
```
