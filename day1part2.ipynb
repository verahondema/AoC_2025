```python
counter = 0
import re
with open('input1.txt', 'r') as file:
    instructions = file.readlines()
    #retrieve all instructions in separate lines
    dial = 50
    #the dial starts at 50
    for instruction in instructions:
        #read every instruction apart
        direction = instruction[0]
        #retrieve the direction from the instruction
        totalPoints = int(instruction[1:])
        #retrieve the amount of points one should turn the dial
        amountOfRounds = int(totalPoints/100)
        #amount that the dial fully turns
        counter = counter + amountOfRounds
        #every time the dial fully turns the counter ticks one
        points = totalPoints - (amountOfRounds * 100)
        #remove all complete rounds the dial turns
        if direction == 'L':
            #when the dial has to turn to the left
            dial = dial - points
            if dial == 0:
                counter = counter + 1
            elif dial < 0:
                if abs(dial) != points:
                    counter = counter + 1
                dial = 100 - abs(dial)
                
        elif direction == 'R':
            dial = dial + points
            if dial == 100:
                counter = counter + 1
                dial = 0
            elif dial > 100:
                dial = dial - 100
                counter = counter + 1

print(f"Counter: {counter}")
```
